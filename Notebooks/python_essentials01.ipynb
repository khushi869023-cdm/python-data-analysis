#### .format ??

```python
# Input Variables 
cost_of_ice_bag = 1.25
profit_margin = .2
number_of_bags = 500 

# Template for output message 
output_template = """If a grocery store sells ice bags at $ {} per bag , with a profit margin of {} % , then the total profit it makes by selling {} ice bags is $ {}"""
print (output_template)
```

```python
# Inserting values into the template
total_profit = number_of_bags * cost_of_ice_bag * profit_margin 
output_message = output_template.format(cost_of_ice_bag, profit_margin*100, number_of_bags, total_profit)
print (output_message)
```

the placeholders {} in the output_template string are replaced with the arguments provided to the .format method

#### Dictionary 
An unodered collection of values . There is no order how we access those values.
Each item in dict. is stored as key:value pairs 
- Key is used to retrieve values from dict. 
- type : dict 
- represented in {} <br>
Dictionary stores many pieces of information eg details about a person , in a single variable. 


```python
person1 = {
    'name': 'Khushi' ,
    'sex' : 'Female' ,
    'age': 23 ,
    'married' : False 
} 
```

```python
person1
```

Dictionary can also be created using dict functions 

```python
person2 = dict(name= 'Tane' , sex='Female', age = 27)
```

```python
type(person1)
```

To access the value in dictionary we do it by providing the key. So we need to know what keys do wewant to access using [ ]  

```python
person1 ['name']
```

If the key isn't present in the dictionary , then a KeyError is returned , to avoid this .get method is used 

```python
person1.get("name")
```

The get method also accepts a default value which is returned if the key is not present in the dictionary 
e.g : "address" ; not present returns 'unknown' if - 
person1.get("address" , "unknown")

```python
person1.get('address','unknown')
```

```python
'name' in  person1 # Checks whether a key is present or not 
```
