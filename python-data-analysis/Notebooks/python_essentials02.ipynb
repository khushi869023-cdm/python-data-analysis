#### Branching and Loops 

#### Branching 
Branching is the ability to make decision and execute a different set of statements based on whether one or more conditions are true. If , else and elif 

```python
if condition: 
    statement1
    statement2
```

Condition can either be a variable or an expression. If True , then statements within the if block are executed. tab space is known as indentation.             

```python
a_number = 34
if a_number % 2 == 0 :
    print("We're inside an if block")
    print('The given number {} is even'.format(a_number))
```

Modulus operator to find the remainder of given no. by 2 <br>
Comparison operator == to compare if remainder is 0 <br>
indicating that the number is divisible by 2 i.e it is even.

#### Shorthand if conditional expression 
A common use case of the if statement involves testing a condition and setting the value of a variable based on the condition. 

```python
a_number = 13

if a_number % 2 == 0 :
    parity= 'even'
else: 
    parity = 'odd'
print('The number {} is {}.'.format(a_number , parity))
```

Python provides a shorter syntax, which allows writing such conditions in a single line of code. It is known as conditional expression, also reffered as ternary operator.<br>

x= true_value if condition else false_value

# Iteration with while loops

Extension of branching : it is the ability to run one or more statements multiple times. This feature is reffered as iteration on looping, and thee are two ways to do this in python : using while loops and for loops 

the statements in block under while are executed repeatedly as long as the condition evaluates to be true 

#### Calculate factorial of 100 using a while loop 
- The factorial of number n is defined as the product of all the numbers from 1 to n <br>
1*2*3*....*(n-2)*(n-1)*n

```python
result = 1
i = 1 #Counter variable or index variable

while i <= 100:
    result = result * i
    i = i + 1 
      
print('The factorial of 100 is {}'.format(result))
```

Here's how thos code works:<br>
We initialize two variables result and i. Result will contain final result of the computation , And i is used to keep track of the next number to be multiplied with the result. <br>
- Both are initialized to 1. 
- The condition i<=100 holds true since i is initially 1 --> statements blocks executed. 
- Result is updated to result * i and i is increased by 1 , now i has value 2 
- i<=100 again holds true for 2 thus while condition was executed again. result * 2 and i+1 = 3. AGAIN TRUE
- Process is repeated till condition becomes false. at 101 and execution stops

### Another example of using two while loops together 

```python
line = '*'
max_length = 10

while len(line) <= max_length:
    print(line)
    line += "*"
while len(line) > 0:
    print(line)
    line = line[:-1]
```

## Break and continue statements 

use break statement within the body of the loop to immediately stop execution and break out of the loop (even if the condition provided to while holds true )

```python
i = 1
result = 1

while i <= 100:
    result = result * i
    if i == 42:
        print ('Magic number 42 reached ! Stopping execution')
        break
    i += 1

    print('i:' , i)
    print('result:',result)
```

## Continue statement
Sometimes we don't want to end the loop entirely , but simply skip the remaining statements in the loop and continue to the next loop 


```python
i = 1 
result

while i < 20:
    i += 1
    if i % 2 == 0:
        print('Skipping {}'.format(i))
        continue    
    print('Multiplying with {}'.format(i))
    result = result * i

print ('i:', i)
print ('result:', result)
```
